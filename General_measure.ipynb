# General measure

## Definition 2.7

A set function $\mu$ on a $\sigma-field$ <font face="cursive">F</font> in $\Omega$ is a measure if

(1) $\mu(A) \in [0, \infty)$ for $A \in$ <font face="cursive">F</font>

(2) $\mu(\emptyset)$ = 0

(3) If $A_1, A_2, ... $ is a disjoint sequence of <font face="cursive">F</font>- set and if $\underset{n=1}{\overset{\infty}{\bigcup}}A_n \in$ <font face="cursive">F</font>$_1$ then $\mu(\underset{n=1}{\overset{\infty}{\bigcup}}A_n) = \sum_{n=1}^{\infty} \mu(A_n)$

## Remark

(1) If $\mu(\Omega) < \infty$, then $\mu$ is called finite.

Particularly, if $\mu(\Omega$) = 1, then $\mu$ is a probability measure.

(2) If $\Omega = A_1 \cup A_2 \cup ... $, for same finite or countable sequence of <font face="cursive">F</font> - sets satisfying $\mu(A_k) < \infty$,

then $\mu$ is $\sigma-finite$

(3) If $\mu$ is a measure on a $\sigma-field$, <font face="cursive">F</font> is in $\Omega$, the triple ($\Omega$, <font face="cursive">F</font>, $\mu$) is a measure space.

E.g 

The probability space will be denoted as ($\Omega$, <font face="cursive">F</font>, <font face="cursive">P</font>) with <font face="cursive">F</font>($\Omega$) = 1 

(4) The pair ($\Omega$, <font face="cursive">F</font>) is measurable space if <font face="cursive">F</font> is a $\sigma-field$

## Theorem 2.4
